```python
x=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
train_data['bedroom_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True),annot=True,cmap='YlGnBu')
```

```python
from sklearn.linear_model import LinearRegression
X_train,y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
test_data=X_test.join(y_train)

test_data['total_rooms']=np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)

test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

test_data['bedroom_ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms']/test_data['households']

X_test,Y_test=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
```

```python
X_train
```

```python
lrg=reg.score(X_test,Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
#scaling the dataset values
x_train_s=scalar.fit_transform(X_train)
x_test_s=scalar.transform(X_test)
reg.fit(x_train_s,y_train)
```

```python
reg.score(x_test_s,Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train_s,Y_train)
```

```python
rfg=rf.score(x_test_s,Y_test)
print(rfg)
```

```python
models = ['Linear Regression', 'Random Forest Regressor']
scores = [lrg*100, rfg*100]

plt.figure(figsize=(10, 5))
bars = plt.bar(models, scores, color=['#4CAF50', '#2196F3'], edgecolor='black', width=0.6)

plt.title("Model Performance Comparison (RÂ² Score)", fontsize=16)
plt.xlabel("Model", fontsize=14)
plt.ylabel("RÂ² Score (%)", fontsize=14)
plt.ylim(0, 100)

for i, score in enumerate(scores):
    plt.text(i, score + 2, f"{score:.2f}%", ha='center', fontsize=12)

plt.show()
```
